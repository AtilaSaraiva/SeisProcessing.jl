```julia
using PyPlot, SeisPlot, SeisMain, SeisProcessing

# FX Deconvolution for SNR enhancement 
```

```julia


d = SeisLinearEvents(dt=0.004,nx1=380,dx1=.4,
                     p1  = [-0.001,0.0003,-0.003],p2=[0.0,0.0,0.0],p3=[0.0,0.0,0.0],
                     p4=[0.0,0.0,0.0],amp=[1.0,-1.0,1.0],
                     tau = [0.6,0.7,1.4])
dn = SeisAddNoise(d,.5,L=5)

Param = Dict(:dt=>0.004,:mu=>0.01,:filter_length=>4,:fmax=>100) 

df = SeisFXDecon(dn,Param);


figure(4,figsize=(6,6))
clf()
SeisPlotTX(hcat(d[:,1:10:end],dn[:,1:10:end],df[:,1:10:end]),style="wiggles",fignum=4)
gcf()
```

```julia
MeasureSNR(d,dn,db=true)
```
