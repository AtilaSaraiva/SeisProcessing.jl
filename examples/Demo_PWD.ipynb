```julia
using PyPlot, SeisPlot, SeisMain, SeisProcessing

# Compute local dips from an image via Plane Wave Destruction

# Sergey Fomel, 2002, Applications of plane-wave destruction filter: Geophysics, 67(6), 1946-1960.
```

```julia
# Prepare data

download("http://seismic.physics.ualberta.ca/data/kevin_0off_velocity.su", "kevin_0off_velocity.su");

SegyToSeis("kevin_0off_velocity.su","kevins_model", format="su", input_type="ieee");

d,h = SeisRead("kevins_model");
```

```julia
coh,pp,res = SeisPWD(d);
```

```julia
SeisPlotTX(d,fignum=1);

# Add quiver plot

L = 1;
pp_x = @. cos((pp-90)*pi/180)*L ;
pp_y = @. sin((pp-90)*pi/180)*L ;

quiver([0:30:350-1],[0:30:600-1],
    pp_x[1:30:end,1:30:end], pp_y[1:30:end,1:30:end],
angles="xy",scale=20,color="purple")
```
